```python
import requests
from bs4 import BeautifulSoup
```

```python
all_packages = []

for page_number in range(1, 355+1):
    URL = "https://community.chocolatey.org/packages?sortOrder=package-title&page=" + str(page_number) + "&prerelease=False&moderatorQueue=False&moderationStatus=all-statuses"
    page = requests.get(URL)
    
    
    soup = BeautifulSoup(page.content, "html.parser")
    
    html_tags = soup.find_all("input", class_="form-control text-bg-theme-elevation-1 user-select-all border-start-0 ps-1")
    for tag in html_tags:
        all_packages.append(tag['value'])
        
    print('Completed page: ' + str(page_number))
```

```python
print(len(all_packages))
```

```python
with open('install_script.txt', mode='wt', encoding='utf-8') as file:
    file.write(' -y\n'.join(all_packages))
```
